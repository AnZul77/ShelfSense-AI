```python
user_counts = ratings_explicit.groupby("User-ID").size()
book_counts = ratings_explicit.groupby("ISBN").size()

filtered_ratings = ratings_explicit[
    ratings_explicit["User-ID"].isin(
        user_counts[user_counts >= 5].index
    )
]

filtered_ratings = filtered_ratings[
    filtered_ratings["ISBN"].isin(
        book_counts[book_counts >= 5].index
    )
]

filtered_ratings.shape
```

```python
filtered_ratings["User-ID"].nunique()
```

```python
filtered_ratings["ISBN"].nunique()
```

```python
user_to_idx = {
    user_id: idx
    for idx, user_id in enumerate(
        filtered_ratings["User-ID"].unique()
    )
}

book_to_idx = {
    isbn: idx
    for idx, isbn in enumerate(
        filtered_ratings["ISBN"].unique()
    )
}
```

```python
filtered_ratings["user_idx"] = (
    filtered_ratings["User-ID"]
    .map(user_to_idx)
)

filtered_ratings["book_idx"] = (
    filtered_ratings["ISBN"]
    .map(book_to_idx)
)
```

```python
filtered_ratings.head()
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    filtered_ratings,
    test_size=0.2,
    random_state=42
)

print("Train:", train_df.shape)
print("Test :", test_df.shape)
```

```python
n_users = train_df["user_idx"].nunique()
n_books = train_df["book_idx"].nunique()

print("Users:", n_users)
print("Books:", n_books)
```

```python
import numpy as np

K = 20  # latent factors

U = np.random.normal(
    scale=0.1,
    size=(n_users, K)
)

V = np.random.normal(
    scale=0.1,
    size=(n_books, K)
)

print(U.shape)
print(V.shape)
```

```python
sample = train_df.iloc[0]

user_idx = sample["user_idx"]
book_idx = sample["book_idx"]
actual_rating = sample["Rating"]

print("User:", user_idx)
print("Book:", book_idx)
print("Actual Rating:", actual_rating)
```

```python
pred_rating = np.dot(
    U[user_idx],
    V[book_idx]
)

print("Predicted:", pred_rating)
```

```python
error = actual_rating - pred_rating

print(error)
```

```python
loss = error ** 2

print(loss)
```

```python
print(U[user_idx][:5])
print(V[book_idx][:5])
```

```python
lr = 0.01
```

```python
u_old = U[user_idx].copy()
v_old = V[book_idx].copy()
```

```python
U[user_idx] += lr * error * v_old
V[book_idx] += lr * error * u_old
```

```python
new_pred = np.dot(
    U[user_idx],
    V[book_idx]
)

print("Old prediction:", pred_rating)
print("New prediction:", new_pred)
```
